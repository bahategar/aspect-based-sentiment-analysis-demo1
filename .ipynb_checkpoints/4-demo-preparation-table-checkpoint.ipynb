```python
import pandas as pd
import numpy as np

import json
```

# Table Format 1

```python
# Load JSON data from a file
with open('data-3.json', 'r') as json_file:
    data = json.load(json_file)

data
```

```python
# Table 1
def get_table_1(input_):
    result = []
    # Looping through level 1 dictionary
    for _, dict_1 in input_.items():
        # Extract the data that needed
        new_data = {k:v for k, v in dict_1.items() if k not in ['aspect', 'topic']}
        new_data.update({k:v for k, v in dict_1['topic'].items() if k != 'term'})
        # Store the data into storage
        result.append(new_data)
        
    return result

table_1 = get_table_1(data)

table_1
```

```python
# Table 2
def get_table_2(input_):
    result = []
    # Looping through level 1 dictionary
    for _, dict_1 in input_.items():
        # Looping through level 2 dictionary
        id = dict_1['id']
        for aspect, dict_2 in dict_1['aspect'].items():
            # Looping through level 3 dictionary
            for label, data in dict_2.items():
                # Looping through level 4 list of data
                for d in data:
                    result.append({
                        'id': id,
                        'aspect': aspect,
                        'type_property': label,
                        'keyword': ', '.join(d[0]).strip(),
                        'part_sentence': d[1],
                        'sentiment': d[2]
                    })
    return result

table_2 = get_table_2(data)
table_2
```
