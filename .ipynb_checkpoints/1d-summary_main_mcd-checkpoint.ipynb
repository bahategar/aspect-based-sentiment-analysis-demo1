```python
def get_nearest_word(sentence, words, model, threshold=0.0, get_all=False):
    
    special_chars = set("?!.,")
    if (not sentence) or (sentence in special_chars):
        return None
    
    def sentence_to_avg_vector(sentence):
        words = sentence.split()
        word_vectors = []
        for word in words:
            if word in model:
                word_vectors.append(model[word])
        
        if not word_vectors:
            return []
        
        avg_vector = np.mean(word_vectors, axis=0)
        return avg_vector
    
    def cosine_similarity(vec1, vec2):
        return 1 - cosine(vec1, vec2)
    
    temp = {}
    current_max, current_word = 0, None
    sentence_vector = sentence_to_avg_vector(sentence)
    
    for word in words:
        if word not in model:
            continue
        word_vector = model[word]
        
        if len(sentence_vector) > 0:
            similarity = cosine_similarity(sentence_vector, word_vector)
        else:
            similarity = 0
        
        if similarity > current_max:
            current_max = similarity
            current_word = word
        
        if similarity >= threshold:
            temp[word] = similarity
    
    
    if get_all:
        temp = dict(sorted(temp.items(), key=lambda x: x[1], reverse=True))
        return temp
    
    if current_max >= threshold:
        return current_word
    
    return
```
